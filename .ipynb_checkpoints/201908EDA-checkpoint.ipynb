# Dropping some columns
We see that some of the columns above are unnecessary at this point, so I will drop them from our dataframe. We might want to use them later however, such as the longitude and latitude data if we go in a geospatial direction (we should!).

```python
extra_columns = ['start station id', 'start station latitude','start station longitude'\
                     ,'end station id', 'end station latitude', 'end station longitude']
df.drop(extra_columns, axis=1, inplace=True)
df.columns
```

## Reformating trip duration
The trip duration column is measured in seconds. Let's change that to be measured in hours.

```python
df['tripduration_hrs'] = df['tripduration']/(60*60)
df.columns
```

### Trip duration histogram in hours

```python
bins=np.linspace(0,2,num=41)

df['tripduration_hrs'].hist(bins=bins)
plt.xlabel('hours')
plt.title('Trip Duration')

plt.savefig('Trip_Duration.png')
```

# Long trips
Let's look at these values where the trip duration is exceedingly long. We'll restrict to situations where the trip duration is over 24 hours. This may indicate the bike was lost and recovered later, or some other situation.

```python
# We'll isolate the instances where 'tripduration' is longer than a day (24*60*60). Then we count the usertypes
# who fall in this range.
df_longtrips=df.loc[df['tripduration_hrs']>24 , ['usertype']]
df_longtrips['usertype'].value_counts()
```

It appears that about 8 times as many customers as subscribers held a bike for longer than a day.

# Convert start and end times to datetime datatype

```python
df['start_time'] = pd.to_datetime(df['start_time'])
```

```python
df['end_time'] = pd.to_datetime(df['end_time'])
df['end_time']
```

# Add days of weeks
0 = Monday,
1 = Tuesday,
2 = Wednesday,
3 = Thursday,
4 = Friday,
5 = Saturday,
6 = Sunday

```python
df['dayofweek'] = df['start_time'].dt.dayofweek
df.tail()
```

```python
df.rename(columns={'dayofweek':'startdayofweek'},inplace=True)
df.head()
```

```python
df['enddayofweek'] = df['end_time'].dt.dayofweek
```

There are at least 72641 - 71330 = 311 instances of a bike being rented on one day and returned on a different day. This does not account for the possibility that a bike is returned one week later.

## Histogram of usage by hour

```python
df["start_time"].dt.hour.hist(bins=24)
plt.title('Usage by hour starting at midnight')
plt.xlabel('hour')
plt.ylabel('count')

plt.savefig('Hourly_Usage.png')
```

## Stacked Bar Graph by user type

```python
userdays=df.groupby(df['dayofweek'])['usertype'].value_counts()
userdays
```

```python
labels=['M','Tu','W','Th','F','Sa','Su']

subscribers=[]
customers=[]
for j in range(0,7):
    subscribers.append(userdays[j]['Subscriber'])
    customers.append(userdays[j]['Customer'])
    
fig, ax =plt.subplots()

ax.bar(labels, subscribers, width=.45, label='Subscribers')
ax.bar(labels, customers, width=.45, label='Customers', bottom=subscribers)
ax.set_ylabel('Usage by user type')
ax.set_title('Usage by day of week')
ax.legend()

#plt.show()

plt.savefig('User_type_dayofweek.png')
```

We see that Friday seems to be the most common day (in this month) and that subscriber usage drops off during the weekend. This lends evidence that the subscribers may be commuters, using the NiceRide bikes on the weekdays.

## Usage by day

```python
days=np.linspace(1,31,31)
riders=df.groupby(df['start_time'].dt.day)['usertype'].count()
plt.plot(days,riders)
plt.title('days in August')
plt.ylabel('rider count')
```
