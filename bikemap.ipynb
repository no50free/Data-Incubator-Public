```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#!conda install -c conda-forge folium=0.5.0 --yes
import folium

```

```python
# import the data set
df = pd.read_csv('/Users/no50free/GitHub/NiceRide/201908-niceride-tripdata.csv')
```

```python
# Grab station names and lat/long. Use only start stations for this.

stations = df[['start station name','start station latitude','start station longitude']]

# There are a lot of NaN station names. Remove those.
stations.dropna(inplace=True)

# Condense dataframe using groupby and mean. There should be an easier/better way to do this, but I think this works
stations = stations.groupby('start station name').mean()
```

### Make the map

```python
mpls_lat_long = [44.9778, -93.2650]

# define the world map
mpls_map = folium.Map(location=mpls_lat_long, zoom_start=12, tiles='Stamen Toner')
# tiles='OpenStreetMap' looks nicer with roads, but renders more slowly.

```

### Populate the map.

```python
# instantiate a feature group for the bike stations in the dataframe
bike_stations = folium.map.FeatureGroup()

# loop through the stations and add each to the feature group
for lat, lng, label in zip(stations['start station latitude'], stations['start station longitude'], stations.index):
    bike_stations.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='blue',
            fill=True,
            fill_color='green',
            fill_opacity=0.6
        )
    )

# add incidents to map
mpls_map.add_child(bike_stations)
```
