```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows = None
from datetime import date, timedelta, datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Use .format(YYYY, M, D)
#lookup_URL = 'https://www.wunderground.com/hourly/us/ny/new-york-city/date/{}-{}-{}.html'
lookup_URL = 'https://www.wunderground.com/history/daily/us/mn/fort-snelling/KMSP/date/{}-{}-{}.html'
start_date = date(2019, 4,22) + pd.Timedelta(days=0)
end_date = date(2019,4,22) + pd.Timedelta(days=222)
df_weather = pd.DataFrame()

options = webdriver.ChromeOptions();
options.add_argument('headless'); # to run chrome in the backbroung

driver = webdriver.Chrome(executable_path='/Users/no50free/GitHub/NiceRide_network_analysis/chromedriver', options=options)

while start_date != end_date:
    print('gathering data from: ', start_date)
    formatted_lookup_URL = lookup_URL.format(start_date.year,
                                             start_date.month,
                                             start_date.day)
    
    driver.get(formatted_lookup_URL)
    rows_rain = WebDriverWait(driver, 25).until(EC.visibility_of_all_elements_located((By.XPATH, '//td[@class="mat-cell cdk-cell cdk-column-precipRate mat-column-precipRate ng-star-inserted"]')))
    rows_time = WebDriverWait(driver, 25).until(EC.visibility_of_all_elements_located((By.XPATH, '//td[@class="mat-cell cdk-cell cdk-column-dateString mat-column-dateString mat-table-sticky ng-star-inserted"]')))
    rows_temp = WebDriverWait(driver, 25).until(EC.visibility_of_all_elements_located((By.XPATH, '//td[@class="mat-cell cdk-cell cdk-column-temperature mat-column-temperature ng-star-inserted"]')))
    rows_dewp = WebDriverWait(driver, 25).until(EC.visibility_of_all_elements_located((By.XPATH, '//td[@class="mat-cell cdk-cell cdk-column-dewPoint mat-column-dewPoint ng-star-inserted"]')))
    rows_wind = WebDriverWait(driver, 25).until(EC.visibility_of_all_elements_located((By.XPATH, '//td[@class="mat-cell cdk-cell cdk-column-windSpeed mat-column-windSpeed ng-star-inserted"]')))
    for row in range(len(rows_rain)):
        time = rows_time[row].find_element_by_xpath('.//span[@class="ng-star-inserted"]').text
        prep = rows_rain[row].find_element_by_xpath('.//span[@class="wu-value wu-value-to"]').text
        temp = rows_temp[row].find_element_by_xpath('.//span[@class="wu-value wu-value-to"]').text
        dewp = rows_dewp[row].find_element_by_xpath('.//span[@class="wu-value wu-value-to"]').text
        wind = rows_wind[row].find_element_by_xpath('.//span[@class="wu-value wu-value-to"]').text
        # append new row to table
        df_weather = df_weather.append(pd.DataFrame({"Day":[str(start_date)],
                                                     "Time":[time],
                                                     "Temperature":[temp],
                                                     "Wind Speed":[wind],
                                                     "Dew Point":[dewp],
                                                     'Precipitation':[prep]}),
                                 ignore_index = True)
    
    start_date += timedelta(days=1)
df_weather
```
